```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM - TEXT ANALYTICS
# TEXTBLOB + NLTK
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS TEXT ANALYTICS?
# --------------------------------------------------------------------------
# Text Analytics (also called Text Mining or NLP - Natural Language Processing)
# is the process of extracting useful information and patterns from text data.
# Computers cannot understand raw text directly, so we process it into
# structured numerical form.
#
# Common tasks in Text Analytics:
#   - Tokenization    : Splitting text into words or sentences
#   - Stopword Removal: Removing common unimportant words
#   - Stemming        : Reducing words to their root form
#   - Lemmatization   : Reducing words to their dictionary base form
#   - POS Tagging     : Labeling each word with its grammatical role
#   - Spell Correction: Fixing spelling mistakes
#   - N-grams         : Sequences of N consecutive words
#   - Term Frequency  : Counting how often each word appears

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# textblob   : High-level NLP library built on top of NLTK.
#              Easy to use for spell correction, tokenization, parsing, etc.
# nltk       : Natural Language Toolkit - the main NLP library in Python.
#              Provides tokenizers, stemmers, lemmatizers, POS taggers, etc.
# matplotlib : For plotting frequency distribution graphs.
# collections.Counter : For counting word frequencies.

# --------------------------------------------------------------------------
# PART 1 - TEXTBLOB
# --------------------------------------------------------------------------

# 1. SPELL CORRECTION
#    b = TextBlob("I havv good spelling")
#    b.correct()
#    - TextBlob automatically detects and corrects spelling mistakes.
#    - 'havv' gets corrected to 'have'.
#    - Uses a built-in dictionary and probability-based correction.

# 2. WORD TOKENIZATION
#    b1.words
#    - Splits the text into individual words (tokens).
#    - Returns a WordList (like a list of strings).
#    - Example: 'Beautiful is better' -> ['Beautiful', 'is', 'better']

# 3. SENTENCE TOKENIZATION
#    b1.sentences
#    - Splits the text into individual sentences.
#    - Returns a list of Sentence objects.

# 4. SINGULARIZE AND PLURALIZE
#    sentence.words[2].singularize()
#    - Converts a plural word to its singular form.
#    - Example: 'spaces' -> 'space'
#
#    sentence.words[5].pluralize()
#    - Converts a singular word to its plural form.
#    - Example: 'level' -> 'levels'
#
#    animals.words.pluralize()
#    - Pluralizes all words in the WordList at once.
#    - Example: ['cat', 'dog', 'octopus'] -> ['cats', 'dogs', 'octopi']

# 5. WORD COUNT
#    sen.word_counts['ekki']
#    - Returns how many times 'ekki' appears (case-insensitive by default).
#
#    sen.words.count('ekki')
#    - Same result using .count() method.
#
#    sen.words.count('ekki', case_sensitive=True)
#    - Case-sensitive count (only counts exact case match).

# 6. PARSING (POS Tagging via TextBlob)
#    b.parse()
#    - Parses the sentence and tags each word with its grammatical role.
#    - Example output: 'And/CC/CC/O now/RB/ADVP/O for/IN/PP/O ...'
#    - Tags: NN=Noun, VB=Verb, JJ=Adjective, RB=Adverb, CC=Conjunction, etc.

# 7. SLICING, UPPER, FIND
#    b1[0:19]  -> Slices the TextBlob like a string.
#    b1.upper() -> Converts all text to uppercase.
#    b1.find('Simple') -> Returns the index (position) of the word 'Simple'.

# 8. COMPARISON
#    apple_blob < banana_blob
#    - Compares two TextBlobs alphabetically (like comparing strings).
#    - 'apples' < 'bananas' -> True (because 'a' comes before 'b').

# 9. N-GRAMS
#    blob.ngrams(n=3)
#    - Returns all sequences of 3 consecutive words (trigrams).
#    - Example: 'Now is better' -> [['Now','is','better'],['is','better','then'],...]
#    - n=2 = bigrams, n=3 = trigrams.
#    - Used in language models and text prediction.

# --------------------------------------------------------------------------
# PART 2 - NLTK
# --------------------------------------------------------------------------

# 10. SENTENCE TOKENIZATION
#     from nltk.tokenize import sent_tokenize
#     tokenized_text = sent_tokenize(text)
#     - Splits a paragraph into individual sentences.
#     - Uses punctuation (. ? !) to detect sentence boundaries.
#     - Returns a list of sentence strings.

# 11. WORD TOKENIZATION
#     from nltk.tokenize import word_tokenize
#     tokenizer_word = word_tokenize(text)
#     - Splits text into individual words AND punctuation marks.
#     - Returns a list of tokens.
#     - Example: 'Good Day!' -> ['Good', 'Day', '!']

# 12. FREQUENCY DISTRIBUTION
#     from nltk.probability import FreqDist
#     fdist = FreqDist(tokenizer_word)
#     - Counts how many times each word/token appears.
#     - fdist.most_common(4) -> Returns the 4 most frequent words.
#     - fdist.plot(30) -> Plots a frequency graph of the top 30 words.

# 13. STOPWORD REMOVAL
#     from nltk.corpus import stopwords
#     stop_words = set(stopwords.words('english'))
#     - Stopwords are common words that carry little meaning:
#       'is', 'the', 'a', 'are', 'in', 'how', 'you', etc.
#     - We remove them to focus on meaningful words.
#     - The loop checks each word: if it's NOT a stopword, keep it.
#     - filtered_sent contains only meaningful words.

# 14. STEMMING
#     from nltk.stem import PorterStemmer
#     ps = PorterStemmer()
#     ps.stem(word)
#     - Stemming cuts off word endings to get the ROOT form.
#     - It is a CRUDE process - the result may not be a real word.
#     - Examples:
#         'flying'  -> 'fli'
#         'running' -> 'run'
#         'studies' -> 'studi'
#     - Fast but not always accurate.

# 15. LEMMATIZATION
#     from nltk.stem.wordnet import WordNetLemmatizer
#     lem = WordNetLemmatizer()
#     lem.lemmatize(word, 'v')  # 'v' means treat word as a verb
#     - Lemmatization reduces a word to its DICTIONARY BASE FORM (lemma).
#     - It is SMARTER than stemming - always returns a real word.
#     - Examples:
#         'flying' (verb) -> 'fly'
#         'better'        -> 'good'
#         'studies'       -> 'study'
#     - Requires WordNet (a large English dictionary database).
#
# STEMMING vs LEMMATIZATION:
#   Stemming    : Fast, crude, may give non-real words (e.g., 'fli')
#   Lemmatization: Slower, accurate, always gives real words (e.g., 'fly')

# 16. POS TAGGING (Part-of-Speech Tagging)
#     tokens = nltk.word_tokenize(sent)
#     nltk.pos_tag(tokens)
#     - Labels each word with its grammatical role.
#     - Common POS Tags:
#         NNP  = Proper Noun (singular)  -> 'Albert', 'Einstein', 'Germany'
#         VBD  = Verb (past tense)       -> 'was', 'born'
#         IN   = Preposition             -> 'in'
#         CD   = Cardinal number         -> '1879'
#         NN   = Noun (common, singular) -> 'word'
#         JJ   = Adjective               -> 'good'
#         RB   = Adverb                  -> 'quickly'
#         CC   = Coordinating conjunction-> 'and', 'but'
#         DT   = Determiner              -> 'the', 'a'
#     - Used in Named Entity Recognition, grammar analysis, etc.

# 17. TERM FREQUENCY
#     from collections import Counter
#     term_frequency = Counter(sentence.split())
#     - Counter counts how many times each word appears in the sentence.
#     - sentence.split() splits the sentence into words by spaces.
#     - Example: 'Texas A&M University is located in Texas'
#       -> Counter({'Texas': 2, 'A&M': 1, 'University': 1, ...})
#     - Term Frequency (TF) is the foundation of text-based ML models
#       like TF-IDF used in search engines and document classification.

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# NLP              : Natural Language Processing - making computers understand text.
# Token            : A single unit of text (word, punctuation, number).
# Tokenization     : Splitting text into tokens (words or sentences).
# Stopwords        : Common words with little meaning (the, is, a, in, are...).
# Stemming         : Cutting word endings to get root (may not be real word).
# Lemmatization    : Reducing word to dictionary base form (always real word).
# POS Tagging      : Labeling each word with its grammatical role (Noun, Verb...).
# N-gram           : Sequence of N consecutive words.
# Bigram           : 2 consecutive words. Trigram = 3 consecutive words.
# FreqDist         : Frequency distribution - count of each word.
# Term Frequency   : How often a word appears in a document.
# TextBlob         : Easy-to-use NLP library for spell check, parsing, etc.
# WordNet          : Large English dictionary database used by lemmatizer.
# PorterStemmer    : A popular stemming algorithm.
# WordNetLemmatizer: Lemmatizer that uses WordNet dictionary.

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is Text Analytics / NLP?
# A: Text Analytics is the process of extracting useful information from
#    unstructured text data. NLP (Natural Language Processing) is the field
#    of making computers understand and process human language.

# Q: What is Tokenization?
# A: Tokenization is the process of splitting text into smaller units called
#    tokens. Sentence tokenization splits into sentences; word tokenization
#    splits into individual words.

# Q: What are Stopwords and why do we remove them?
# A: Stopwords are common words like 'is', 'the', 'a', 'in' that appear
#    frequently but carry little meaning. We remove them to focus on
#    important words and reduce noise in the data.

# Q: What is the difference between Stemming and Lemmatization?
# A: Stemming cuts word endings to get the root - it is fast but crude
#    and may produce non-real words (e.g., 'flying' -> 'fli').
#    Lemmatization uses a dictionary to find the proper base form - it is
#    slower but always returns a real word (e.g., 'flying' -> 'fly').

# Q: What is POS Tagging?
# A: POS (Part-of-Speech) Tagging labels each word in a sentence with its
#    grammatical role: Noun (NN), Verb (VB), Adjective (JJ), Adverb (RB), etc.

# Q: What is an N-gram?
# A: An N-gram is a sequence of N consecutive words from a text.
#    Bigram = 2 words, Trigram = 3 words.
#    Example: 'Now is better' is a trigram from 'Now is better then never'.

# Q: What is Term Frequency?
# A: Term Frequency (TF) counts how many times each word appears in a document.
#    It is calculated using Counter(sentence.split()).
#    It is the basis of TF-IDF used in search engines.

# Q: What does FreqDist do?
# A: FreqDist (Frequency Distribution) from NLTK counts how many times each
#    word/token appears in the text. most_common(n) returns the top n words.
#    fdist.plot() shows a frequency graph.

# Q: What is TextBlob used for?
# A: TextBlob is a high-level NLP library that makes common tasks easy:
#    spell correction (.correct()), word/sentence tokenization (.words, .sentences),
#    singularize/pluralize, parsing, n-grams, and word counting.

# Q: What is the difference between word_counts and count() in TextBlob?
# A: word_counts['word'] is case-insensitive by default.
#    .count('word') is also case-insensitive by default, but you can pass
#    case_sensitive=True to make it case-sensitive.
# =============================================================================
```
