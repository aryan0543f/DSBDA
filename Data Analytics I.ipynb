## Reading Required Dataset

```python
from sklearn.datasets import fetch_california_housing
boston = fetch_california_housing()
```

```python
print(boston)
```

## # Importing Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

## # Creating Dataframe

```python
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
df_y = pd.DataFrame(boston.target)
```

## # Initializing the linear regression Model

```python
reg = linear_model.LinearRegression()
```

## # Splitting Data into 70% training and 30% testing

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=4)
```

## # Train the model

```python
reg.fit(x_train, y_train)
```

```python
#print the coefficient/ weights for each feature/column
#f(x) = mx + c =y where m is the coefficient

print(reg.coef_)
```

## # Prediction

```python
#print prediction for model
y_pred = reg.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
#check the model accuracy using mean squared error
print(np.mean((y_pred-y_test)**2))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM - DATA ANALYTICS I
# LINEAR REGRESSION
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS THIS PRACTICAL ABOUT?
# --------------------------------------------------------------------------
# This practical is about Linear Regression - a supervised machine learning
# algorithm used to PREDICT a continuous numerical value.
# We use the California Housing dataset to predict house prices based on
# various features like number of rooms, population, location, etc.

# --------------------------------------------------------------------------
# WHAT IS LINEAR REGRESSION?
# --------------------------------------------------------------------------
# Linear Regression finds the best straight line (or hyperplane in multiple
# dimensions) that fits the data.
#
# Formula: y = mx + c
#   y = predicted output (house price)
#   x = input features (rooms, location, etc.)
#   m = coefficient (slope / weight) - how much y changes per unit of x
#   c = intercept - value of y when x = 0
#
# When there are multiple input features (like here), it is called
# MULTIPLE LINEAR REGRESSION:
#   y = m1*x1 + m2*x2 + ... + mn*xn + c

# --------------------------------------------------------------------------
# DATASET USED - CALIFORNIA HOUSING DATASET
# --------------------------------------------------------------------------
# Source  : sklearn.datasets (fetch_california_housing)
# About   : Contains data about housing in California districts.
# Target  : Median house value (in $100,000s) - what we want to predict.
#
# Features (8 input columns):
#   MedInc      : Median income of households in the block
#   HouseAge    : Median age of houses in the block
#   AveRooms    : Average number of rooms per household
#   AveBedrms   : Average number of bedrooms per household
#   Population  : Total population of the block
#   AveOccup    : Average number of household members
#   Latitude    : Geographic latitude of the block
#   Longitude   : Geographic longitude of the block
#
# Shape: 20640 rows x 8 columns (features)

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# pandas (pd)                        : For creating and managing DataFrames.
# numpy  (np)                        : For numerical operations (e.g., mean).
# sklearn.linear_model               : Provides LinearRegression model.
# sklearn.model_selection            : Provides train_test_split function.
# sklearn.metrics.mean_squared_error : For evaluating model accuracy.

# --------------------------------------------------------------------------
# STEP 1 - LOADING THE DATASET
# --------------------------------------------------------------------------
# fetch_california_housing() : Loads the dataset from sklearn.
#   boston.data               : The feature matrix (input values).
#   boston.target             : The target array (house prices to predict).
#   boston.feature_names      : Names of the 8 feature columns.
#
# df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
#   -> Creates a DataFrame for input features (X).
# df_y = pd.DataFrame(boston.target)
#   -> Creates a DataFrame for the target/output (Y).

# --------------------------------------------------------------------------
# STEP 2 - EXPLORING THE DATA
# --------------------------------------------------------------------------
# df_x.describe() : Shows statistics (mean, std, min, max, etc.) for features.
# df_y.describe() : Shows statistics for the target (house prices).
# df_x.shape      : Returns (20640, 8) - 20640 samples, 8 features.
# df_x.columns    : Lists all 8 feature names.

# --------------------------------------------------------------------------
# STEP 3 - INITIALIZING THE MODEL
# --------------------------------------------------------------------------
# reg = linear_model.LinearRegression()
# - Creates a Linear Regression model object.
# - At this point the model is empty - it has not learned anything yet.

# --------------------------------------------------------------------------
# STEP 4 - SPLITTING DATA INTO TRAIN AND TEST
# --------------------------------------------------------------------------
# train_test_split(df_x, df_y, test_size=0.3, random_state=4)
#
# - Splits the dataset into two parts:
#     Training set (70%) : Used to TRAIN (teach) the model.
#     Testing set  (30%) : Used to TEST (evaluate) the model on unseen data.
#
# - test_size=0.3  : 30% of data goes to testing, 70% to training.
# - random_state=4 : Fixes the random seed so the split is reproducible
#                    (same split every time you run the code).
#
# Returns 4 variables:
#   x_train : Training features
#   x_test  : Testing features
#   y_train : Training labels (actual prices for training)
#   y_test  : Testing labels  (actual prices for testing/comparison)

# --------------------------------------------------------------------------
# STEP 5 - TRAINING THE MODEL
# --------------------------------------------------------------------------
# reg.fit(x_train, y_train)
# - Trains the model using the training data.
# - The model learns the best values of m (coefficients) and c (intercept)
#   that minimize the error between predicted and actual house prices.
# - This is the LEARNING step.
#
# reg.coef_
# - After training, this gives the coefficient (weight/slope) for each feature.
# - Each coefficient tells how much the house price changes when that
#   feature increases by 1 unit.
# - Positive coefficient -> feature increases price.
# - Negative coefficient -> feature decreases price.

# --------------------------------------------------------------------------
# STEP 6 - MAKING PREDICTIONS
# --------------------------------------------------------------------------
# y_pred = reg.predict(x_test)
# - Uses the trained model to predict house prices for the test data.
# - x_test contains features the model has NEVER seen before.
# - y_pred contains the model's predicted prices.
# - We compare y_pred with y_test (actual prices) to check accuracy.

# --------------------------------------------------------------------------
# STEP 7 - EVALUATING THE MODEL (Mean Squared Error)
# --------------------------------------------------------------------------
# METHOD 1 - Manual calculation:
#   np.mean((y_pred - y_test)**2)
#   - Calculates the average of squared differences between predicted
#     and actual values.
#
# METHOD 2 - Using sklearn:
#   mean_squared_error(y_test, y_pred)
#   - Same result, but using sklearn's built-in function.
#
# WHAT IS MSE?
#   MSE = Mean Squared Error
#   Formula: MSE = (1/n) * sum((actual - predicted)^2)
#   - Measures how far the predictions are from the actual values.
#   - Lower MSE = better model (predictions are closer to actual values).
#   - Higher MSE = worse model (predictions are far from actual values).
#   - Squaring the errors penalizes large errors more heavily.
#   - Here MSE ≈ 0.52 which means predictions are reasonably close.

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# Supervised Learning    : Model learns from labeled data (input + correct output).
# Linear Regression      : Predicts a continuous value using a straight line.
# Multiple Regression    : Linear regression with more than one input feature.
# Coefficient (m)        : Weight/slope - how much output changes per unit input.
# Intercept (c)          : Value of output when all inputs are zero.
# Training Data          : Data used to teach/train the model.
# Testing Data           : Unseen data used to evaluate model performance.
# Overfitting            : Model performs well on training but poorly on testing.
# MSE                    : Mean Squared Error - measures prediction error.
# reg.fit()              : Trains the model.
# reg.predict()          : Makes predictions using the trained model.
# reg.coef_              : Learned coefficients (weights) for each feature.
# train_test_split()     : Splits data into training and testing sets.
# random_state           : Fixes randomness so results are reproducible.

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is Linear Regression?
# A: Linear Regression is a supervised machine learning algorithm that
#    predicts a continuous numerical output by finding the best-fit line
#    through the data. Formula: y = mx + c.

# Q: What dataset is used and what are we predicting?
# A: California Housing dataset from sklearn. We are predicting the median
#    house value (price) based on 8 features like income, rooms, location.

# Q: Why do we split data into training and testing sets?
# A: Training set is used to teach the model. Testing set is used to check
#    how well the model performs on new, unseen data. This prevents overfitting
#    and gives a fair evaluation of the model.

# Q: What is test_size=0.3 and random_state=4?
# A: test_size=0.3 means 30% of data is used for testing and 70% for training.
#    random_state=4 fixes the random seed so the same split happens every time.

# Q: What does reg.fit() do?
# A: It trains the model by learning the best coefficients (m) and intercept (c)
#    from the training data to minimize prediction error.

# Q: What does reg.coef_ give us?
# A: It gives the coefficient (weight) for each input feature.
#    It tells how much the house price changes when that feature increases by 1.

# Q: What is Mean Squared Error (MSE)?
# A: MSE measures the average squared difference between predicted and actual
#    values. Lower MSE means better predictions.
#    Formula: MSE = (1/n) * sum((actual - predicted)^2)

# Q: What is the difference between y_test and y_pred?
# A: y_test = actual house prices from the dataset (ground truth).
#    y_pred = prices predicted by our trained model.
#    We compare them using MSE to measure model accuracy.

# Q: What is Supervised vs Unsupervised Learning?
# A: Supervised Learning: Model is trained on labeled data (input + output).
#    Example: Linear Regression, Classification.
#    Unsupervised Learning: Model finds patterns in data without labels.
#    Example: Clustering (K-Means).
# =============================================================================
```
