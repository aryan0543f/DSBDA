```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM - DATA VISUALIZATION II
# ADVANCED SEABORN PLOTS ON TITANIC DATASET
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS THIS PRACTICAL ABOUT?
# --------------------------------------------------------------------------
# This practical focuses on advanced data visualization using seaborn and
# pandas on the Titanic dataset.
# We explore survival patterns based on gender and age using various plots
# like countplot, pie chart, barplot, boxplot, heatmap, and clustermap.

# --------------------------------------------------------------------------
# DATASET USED - TITANIC DATASET
# --------------------------------------------------------------------------
# Key columns used in this practical:
#   Sex      : Gender of the passenger (male / female)
#   Survived : Whether the passenger survived (0 = No, 1 = Yes)
#   Age      : Age of the passenger
#   Cabin    : Cabin number (missing values replaced with 'unknown')
#   Embarked : Port of embarkation (missing values filled with mode)

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# pandas (pd)       : For loading data and creating crosstabs.
# numpy  (np)       : For numerical operations (np.nan for missing values).
# matplotlib.pyplot : Base plotting library.
# seaborn (sns)     : High-level statistical visualization library.
# %matplotlib inline: Shows plots inside the Jupyter notebook.

# --------------------------------------------------------------------------
# DATA PREPROCESSING
# --------------------------------------------------------------------------
# Same as Data Visualization I:
#   - Cabin   : Replace NaN with 'unknown' using replace()
#   - Age     : Fill missing values using interpolate()
#   - Embarked: Fill missing values with mode (most common port)

# --------------------------------------------------------------------------
# PLOTS USED IN THIS PRACTICAL
# --------------------------------------------------------------------------

# 1. sns.countplot() - Count Plot
#    sns.countplot(x=df['Sex'])
#    sns.countplot(x=df['Survived'])
#
#    - Shows the COUNT (number of rows) for each category.
#    - sns.countplot(x=df['Sex'])      -> How many males vs females on Titanic.
#    - sns.countplot(x=df['Survived']) -> How many survived (1) vs not (0).
#    - x= specifies the column to count on the x-axis.

# 2. Pie Chart (using pandas .plot())
#    df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
#    df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
#
#    - A pie chart shows the proportion of each category as a slice.
#    - value_counts() : Counts how many times each value appears.
#    - .plot(kind='pie') : Plots the counts as a pie chart.
#    - autopct='%.2f'  : Shows the percentage inside each slice
#                        with 2 decimal places (e.g., 64.76%).
#    - Useful for showing percentage breakdown of categories.

# 3. sns.barplot() - Bar Plot
#    sns.barplot(x=df['Survived'], y=df['Age'])
#    sns.barplot(x=df['Sex'], y=df['Survived'])
#    sns.barplot(x=df['Sex'], y=df['Age'], hue=df['Survived'])
#
#    - Shows the AVERAGE of y for each category in x.
#    - Error bars show the confidence interval (uncertainty of the mean).
#    - barplot(x='Survived', y='Age')     -> Average age of survivors vs non-survivors.
#    - barplot(x='Sex', y='Survived')     -> Average survival rate for males vs females.
#    - barplot(x='Sex', y='Age', hue='Survived') -> Average age by gender AND survival.
#    - hue= : Adds a third variable by splitting each bar into sub-groups.
#             Creates grouped bars with different colors per hue value.

# 4. sns.boxplot() - Box Plot
#    sns.boxplot(x=df['Sex'], y=df['Age'])
#    sns.boxplot(x=df['Sex'], y=df['Age'], hue=df['Survived'])
#
#    - Shows the distribution of a numerical variable (y) for each category (x).
#    - The box shows: Q1 (25%), median (50%), Q3 (75%).
#    - Whiskers extend to min and max (within 1.5*IQR).
#    - Dots beyond whiskers are OUTLIERS.
#    - boxplot(x='Sex', y='Age')                  -> Age distribution for males vs females.
#    - boxplot(x='Sex', y='Age', hue='Survived')  -> Age distribution by gender AND survival.
#    - hue= : Splits each box into sub-groups (survived vs not survived).

# 5. pd.crosstab() - Cross Tabulation
#    pd.crosstab(df['Sex'], df['Survived'])
#    pd.crosstab(df['Age'], df['Survived'])
#
#    - Creates a frequency table showing how two categorical variables relate.
#    - Rows = unique values of first variable (Sex: male/female).
#    - Columns = unique values of second variable (Survived: 0/1).
#    - Each cell = count of rows where both conditions are true.
#    - Example output:
#        Survived   0    1
#        Sex
#        female    81  233
#        male     468  109
#    - This is the foundation for heatmap and clustermap.

# 6. sns.heatmap() - Heatmap
#    sns.heatmap(pd.crosstab(df['Sex'], df['Survived']))
#    sns.heatmap(pd.crosstab(df['Age'], df['Survived']))
#
#    - Visualizes a 2D table (like crosstab) as a color-coded grid.
#    - Darker/brighter color = higher value in that cell.
#    - Makes it easy to spot which combinations have high/low counts.
#    - Can add annot=True to show numbers inside each cell.

# 7. sns.clustermap() - Cluster Map
#    sns.clustermap(pd.crosstab(df['Sex'], df['Survived']))
#    sns.clustermap(pd.crosstab(df['Age'], df['Survived']))
#
#    - Like a heatmap BUT also performs hierarchical clustering.
#    - Reorders rows and columns so that similar ones are grouped together.
#    - The dendrograms (tree diagrams) on the sides show the clustering.
#    - Useful for finding natural groupings in data.
#
# DIFFERENCE BETWEEN HEATMAP AND CLUSTERMAP:
#   heatmap    : Shows data in original row/column order. No reordering.
#   clustermap : Reorders rows and columns by similarity using clustering.
#                Adds dendrograms to show the clustering hierarchy.

# --------------------------------------------------------------------------
# SUMMARY TABLE - ALL PLOTS
# --------------------------------------------------------------------------
# | Plot        | Function                    | What it shows                        |
# |-------------|----------------------------|--------------------------------------|
# | countplot   | sns.countplot()            | Count of each category               |
# | pie chart   | .plot(kind='pie')          | Percentage of each category          |
# | barplot     | sns.barplot()              | Average of y per category in x       |
# | boxplot     | sns.boxplot()              | Distribution, median, outliers       |
# | crosstab    | pd.crosstab()              | Frequency table of two variables     |
# | heatmap     | sns.heatmap(crosstab)      | Color-coded crosstab grid            |
# | clustermap  | sns.clustermap(crosstab)   | Heatmap + hierarchical clustering    |

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# value_counts()      : Counts occurrences of each unique value in a column.
# autopct             : Format string for percentage labels in pie chart.
# hue                 : Adds a third variable by splitting into color-coded groups.
# Crosstab            : Frequency table showing relationship between two variables.
# Heatmap             : Color-coded grid visualization of a 2D table.
# Clustermap          : Heatmap with hierarchical clustering and dendrograms.
# Dendrogram          : Tree diagram showing how items are clustered together.
# IQR                 : Interquartile Range = Q3 - Q1 (used in boxplot whiskers).
# Outlier             : Data point far outside the normal range (shown as dots).
# Confidence Interval : Range within which the true mean likely falls (error bars).

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is the purpose of this practical?
# A: To visualize survival patterns in the Titanic dataset using various
#    seaborn and pandas plots, exploring how gender and age relate to survival.

# Q: What does pd.crosstab() do?
# A: It creates a frequency table showing how two categorical variables relate.
#    Each cell shows the count of rows where both conditions are true.
#    Example: how many male/female passengers survived or did not survive.

# Q: What is the difference between heatmap and clustermap?
# A: heatmap shows data in original order as a color-coded grid.
#    clustermap reorders rows and columns by similarity using hierarchical
#    clustering, and adds dendrograms to show the clustering structure.

# Q: What does hue= do in barplot and boxplot?
# A: hue= adds a third variable by splitting each bar/box into sub-groups
#    with different colors. For example, hue='Survived' splits each gender
#    bar into survived (1) and not survived (0) groups.

# Q: What does autopct='%.2f' do in a pie chart?
# A: It shows the percentage of each slice inside the pie chart,
#    formatted to 2 decimal places (e.g., 64.76%).

# Q: What is the difference between barplot and countplot?
# A: countplot shows the COUNT (number of rows) for each category.
#    barplot shows the AVERAGE of a numerical variable for each category.

# Q: What does a boxplot show?
# A: A boxplot shows the distribution of a numerical variable:
#    - Box: Q1 (25%), median (50%), Q3 (75%)
#    - Whiskers: min and max within 1.5*IQR
#    - Dots: outliers beyond the whiskers

# Q: What does value_counts() do?
# A: It counts how many times each unique value appears in a column.
#    Example: df['Sex'].value_counts() -> male: 577, female: 314

# Q: What insights can we get from sns.barplot(x='Sex', y='Survived')?
# A: It shows the average survival rate for males vs females.
#    Since Survived is 0 or 1, the average = proportion who survived.
#    This reveals that females had a much higher survival rate than males.
# =============================================================================
```
