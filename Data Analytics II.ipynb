## Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Data Set/Social_Network_Ads.csv')
df
```

```python
x = df.iloc[:, [2, 3]].values
x
```

```python
y = df.iloc[:, 4].values
y
```

## # Spliting Data into 75% training and 25% testing data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
x_train
```

```python
x_test.shape
```

```python
y_test
```

```python
y_test.shape
```

## # Data Standardization

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)
x_train
```

## # Initializing logistic Regression model and training the model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

## # Prediction

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

## # Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#The confusion matrix tells us that there were 87 correct predictions and 13 incorrect predictions
```

## # Calculating accuracy of model

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
accuracy
```

## # True Positive : Actually True, Predicted True

```python
tp = cm[0,[0]]
print('True Positive : ',tp)
```

## # False Positive : Actually False, Predicted True

```python
fp = cm[0, [1]]
print('False Positive : ',fp)
```

## # False Negative : Actually True, Predicted False

```python
fn = cm[1, [0]]
print('False Negative : ',fn)
```

## # True Negative : Actually False, Predicted False

```python
tn = cm[1, [1]]
print('True Negative : ',tn)
```

## # Accuracy

```python
accuracy_cm = ((tp + tn)/(tp + fp + fn + tn))
print('Accuracy : ',accuracy_cm*100)
```

## # Error Rate

```python
error_rate_cm = ((fp + fn)/(tp + fp + fn + tn))
print('Error Rate : ', error_rate_cm*100)
```

## # Precision

```python
precision_cm = (tp / (tp + fp))
print('Precision : ',precision_cm*100)
```

## # Recall

```python
#recall is also called sensitivity
recall_cm = (tp / (tp + fn))
print('Sensitivity : ', recall_cm*100)
```

## # Specificity

```python
specificity_cm = (tn / (tn + fp))
print('Specificity : ', specificity_cm*100)
```

```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM - DATA ANALYTICS II
# LOGISTIC REGRESSION & CONFUSION MATRIX
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS THIS PRACTICAL ABOUT?
# --------------------------------------------------------------------------
# This practical is about Logistic Regression - a supervised machine learning
# algorithm used for CLASSIFICATION (predicting a category/class).
# We use the Social Network Ads dataset to predict whether a user will
# purchase a product (1 = Yes, 0 = No) based on their Age and Salary.

# --------------------------------------------------------------------------
# WHAT IS LOGISTIC REGRESSION?
# --------------------------------------------------------------------------
# Despite the name, Logistic Regression is used for CLASSIFICATION, not regression.
# It predicts the PROBABILITY that an input belongs to a class (0 or 1).
#
# It uses the Sigmoid Function to convert any value into a probability (0 to 1):
#   sigmoid(z) = 1 / (1 + e^(-z))
#
# Decision Rule:
#   If probability >= 0.5 -> predict class 1 (Yes / Purchased)
#   If probability <  0.5 -> predict class 0 (No  / Not Purchased)
#
# Difference from Linear Regression:
#   Linear Regression  -> predicts a continuous number (e.g., house price)
#   Logistic Regression -> predicts a category/class (e.g., Yes or No)

# --------------------------------------------------------------------------
# DATASET USED - SOCIAL NETWORK ADS DATASET
# --------------------------------------------------------------------------
# Contains data about users of a social network and whether they bought a product.
#
# Columns:
#   User ID    : Unique ID of the user (not used)
#   Gender     : Male or Female (not used)
#   Age        : Age of the user  -> used as input feature (x)
#   Salary     : Estimated salary -> used as input feature (x)
#   Purchased  : 0 = Did not buy, 1 = Bought -> this is what we predict (y)
#
# Shape: 400 rows x 5 columns

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# pandas (pd)                         : For loading and exploring data.
# numpy  (np)                         : For numerical operations.
# matplotlib.pyplot (plt)             : For plotting graphs.
# sklearn.model_selection             : train_test_split to split data.
# sklearn.preprocessing.StandardScaler: For feature scaling/standardization.
# sklearn.linear_model.LogisticRegression : The classification model.
# sklearn.metrics.confusion_matrix    : To evaluate model performance.
# sklearn.metrics.accuracy_score      : To calculate accuracy.

# --------------------------------------------------------------------------
# STEP 1 - SELECTING FEATURES AND TARGET
# --------------------------------------------------------------------------
# x = df.iloc[:, [2, 3]].values
#   - Selects columns at index 2 and 3 (Age and Salary) as input features.
#   - iloc = integer location based indexing.
#   - [:, [2,3]] means all rows, columns 2 and 3.
#   - .values converts DataFrame to a numpy array.
#
# y = df.iloc[:, 4].values
#   - Selects column at index 4 (Purchased) as the target/output.
#   - This is what we want to predict (0 or 1).

# --------------------------------------------------------------------------
# STEP 2 - SPLITTING DATA INTO TRAIN AND TEST
# --------------------------------------------------------------------------
# train_test_split(x, y, test_size=0.25, random_state=0)
#   - test_size=0.25 : 25% for testing, 75% for training.
#   - random_state=0 : Fixes the random seed for reproducibility.
#   - Returns: x_train, x_test, y_train, y_test

# --------------------------------------------------------------------------
# STEP 3 - DATA STANDARDIZATION (Feature Scaling)
# --------------------------------------------------------------------------
# StandardScaler() standardizes features to have mean=0 and std=1.
#
# WHY DO WE SCALE?
#   Age ranges from ~18 to 60, but Salary ranges from ~15000 to 150000.
#   Without scaling, Salary dominates the model because its values are much larger.
#   Scaling brings both features to the same range so neither dominates.
#
# sc_x = StandardScaler()
# x_train = sc_x.fit_transform(x_train)
#   - fit_transform() : Learns the mean and std from training data,
#                       then transforms it.
# x_test = sc_x.fit_transform(x_test)
#   - Transforms test data using the same scaler.
#
# Formula: z = (x - mean) / std

# --------------------------------------------------------------------------
# STEP 4 - TRAINING THE MODEL
# --------------------------------------------------------------------------
# classifier = LogisticRegression(random_state=0)
#   - Creates a Logistic Regression model.
#
# classifier.fit(x_train, y_train)
#   - Trains the model on the training data.
#   - The model learns the relationship between Age/Salary and Purchase decision.

# --------------------------------------------------------------------------
# STEP 5 - MAKING PREDICTIONS
# --------------------------------------------------------------------------
# y_pred = classifier.predict(x_test)
#   - Predicts whether each user in the test set purchased or not (0 or 1).
#   - We compare y_pred with y_test to evaluate the model.

# --------------------------------------------------------------------------
# STEP 6 - CONFUSION MATRIX (Most Important Part)
# --------------------------------------------------------------------------
# cm = confusion_matrix(y_test, y_pred)
#
# A Confusion Matrix is a 2x2 table that shows:
#
#                    PREDICTED
#                  |  0 (No)  |  1 (Yes) |
#   ACTUAL  0 (No) |    TP    |    FP    |
#           1 (Yes)|    FN    |    TN    |
#
# TP (True Positive)  : cm[0][0] - Actually No,  Predicted No  (Correct)
# FP (False Positive) : cm[0][1] - Actually No,  Predicted Yes (Wrong)
# FN (False Negative) : cm[1][0] - Actually Yes, Predicted No  (Wrong)
# TN (True Negative)  : cm[1][1] - Actually Yes, Predicted Yes (Correct)
#
# Correct predictions   = TP + TN
# Incorrect predictions = FP + FN

# --------------------------------------------------------------------------
# STEP 7 - PERFORMANCE METRICS (Calculated from Confusion Matrix)
# --------------------------------------------------------------------------

# 1. ACCURACY
#    accuracy = (TP + TN) / (TP + FP + FN + TN)
#    - Percentage of total correct predictions.
#    - Also calculated using: accuracy_score(y_test, y_pred) * 100

# 2. ERROR RATE
#    error_rate = (FP + FN) / (TP + FP + FN + TN)
#    - Percentage of total wrong predictions.
#    - Error Rate = 1 - Accuracy

# 3. PRECISION
#    precision = TP / (TP + FP)
#    - Out of all predicted positives, how many were actually positive?
#    - High precision = fewer false alarms.

# 4. RECALL (Sensitivity)
#    recall = TP / (TP + FN)
#    - Out of all actual positives, how many did we correctly predict?
#    - Also called Sensitivity or True Positive Rate.
#    - High recall = fewer missed positives.

# 5. SPECIFICITY
#    specificity = TN / (TN + FP)
#    - Out of all actual negatives, how many did we correctly predict as negative?
#    - Also called True Negative Rate.

# --------------------------------------------------------------------------
# SUMMARY TABLE - ALL METRICS
# --------------------------------------------------------------------------
# | Metric       | Formula                        | What it means                    |
# |--------------|--------------------------------|----------------------------------|
# | Accuracy     | (TP+TN)/(TP+FP+FN+TN)         | Overall correct predictions      |
# | Error Rate   | (FP+FN)/(TP+FP+FN+TN)         | Overall wrong predictions        |
# | Precision    | TP/(TP+FP)                     | Correctness of positive preds    |
# | Recall       | TP/(TP+FN)                     | Coverage of actual positives     |
# | Specificity  | TN/(TN+FP)                     | Coverage of actual negatives     |

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# Classification     : Predicting a category (Yes/No, 0/1, Spam/Not Spam).
# Logistic Regression: Classification algorithm using sigmoid function.
# Sigmoid Function   : Converts any value to a probability between 0 and 1.
# StandardScaler     : Scales features to mean=0, std=1.
# fit_transform()    : Learns parameters from data AND transforms it.
# transform()        : Only transforms using already-learned parameters.
# Confusion Matrix   : Table showing TP, FP, FN, TN counts.
# TP                 : True Positive  - correctly predicted positive.
# TN                 : True Negative  - correctly predicted negative.
# FP                 : False Positive - wrongly predicted positive (Type I error).
# FN                 : False Negative - wrongly predicted negative (Type II error).
# Precision          : How accurate are positive predictions?
# Recall/Sensitivity : How many actual positives were caught?
# Specificity        : How many actual negatives were correctly identified?
# iloc               : Integer location indexing in pandas.
# random_state       : Fixes randomness for reproducibility.

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is Logistic Regression?
# A: Logistic Regression is a supervised machine learning algorithm used for
#    binary classification. It predicts the probability of an input belonging
#    to a class (0 or 1) using the sigmoid function.

# Q: What is the difference between Linear and Logistic Regression?
# A: Linear Regression predicts a continuous number (e.g., price, temperature).
#    Logistic Regression predicts a class/category (e.g., Yes/No, 0/1).

# Q: What dataset is used and what are we predicting?
# A: Social Network Ads dataset. We predict whether a user will purchase a
#    product (1=Yes, 0=No) based on their Age and Estimated Salary.

# Q: Why do we use StandardScaler?
# A: Because Age and Salary have very different ranges. Without scaling,
#    Salary (large values) would dominate the model. StandardScaler brings
#    both features to the same scale (mean=0, std=1).

# Q: What is a Confusion Matrix?
# A: A 2x2 table that shows how many predictions were correct and incorrect.
#    It contains TP, FP, FN, TN values used to calculate all performance metrics.

# Q: What is the difference between Precision and Recall?
# A: Precision = TP/(TP+FP) - of all predicted positives, how many are correct?
#    Recall    = TP/(TP+FN) - of all actual positives, how many did we find?

# Q: What is Specificity?
# A: Specificity = TN/(TN+FP) - of all actual negatives, how many did we
#    correctly predict as negative? Also called True Negative Rate.

# Q: What is the difference between Recall and Specificity?
# A: Recall measures performance on actual POSITIVES (TP rate).
#    Specificity measures performance on actual NEGATIVES (TN rate).

# Q: What does accuracy_score() do?
# A: It calculates the percentage of correct predictions.
#    accuracy = (TP + TN) / total predictions * 100

# Q: What is the sigmoid function?
# A: sigmoid(z) = 1 / (1 + e^(-z))
#    It converts any real number into a value between 0 and 1 (probability).
#    If output >= 0.5, predict class 1. If < 0.5, predict class 0.
# =============================================================================
```
