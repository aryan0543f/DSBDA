```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM
# DESCRIPTIVE STATISTICS - MEASURES OF CENTRAL TENDENCY AND VARIABILITY
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS DESCRIPTIVE STATISTICS?
# --------------------------------------------------------------------------
# Descriptive Statistics is a branch of statistics that summarizes and
# describes the main features of a dataset.
# It does NOT make predictions - it just describes what the data looks like.
# It is divided into two main parts:
#   1. Measures of Central Tendency  -> Where is the center of the data?
#   2. Measures of Variability       -> How spread out is the data?

# --------------------------------------------------------------------------
# DATASET USED - IRIS DATASET
# --------------------------------------------------------------------------
# The Iris dataset is one of the most famous datasets in data science.
# It contains measurements of 150 iris flowers from 3 different species:
#   - Iris Setosa
#   - Iris Versicolor
#   - Iris Virginica
#
# Columns (after renaming):
#   - SepalLengthCm : Length of the sepal in centimeters
#   - SepalWidthCm  : Width of the sepal in centimeters
#   - PetalLengthCm : Length of the petal in centimeters
#   - PetalWidthCm  : Width of the petal in centimeters
#   - Species       : Type of iris flower (categorical)
#
# Shape: 150 rows x 5 columns
# No missing values in this dataset.

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# pandas (pd) : For loading and analyzing data using DataFrames.
# numpy  (np) : For numerical operations like min and max.

# --------------------------------------------------------------------------
# DATA PREPROCESSING
# --------------------------------------------------------------------------
# df.shape          : Returns (150, 5) - 150 rows and 5 columns.
# df.isnull().sum() : Checks for missing values - all zeros (no missing data).
# df.dtypes         : Shows data types - 4 float columns + 1 object (Species).
# df.info()         : Summary of DataFrame - column names, types, non-null count.
# df.columns = [...]: Renames columns to standard readable names.
# df.head()         : Shows first 5 rows to verify the data looks correct.

# --------------------------------------------------------------------------
# MEASURES OF CENTRAL TENDENCY
# (These tell us where the CENTER or MIDDLE of the data is)
# --------------------------------------------------------------------------

# 1. MEAN (Average)
#    df.mean(numeric_only=True)
#    - Adds all values and divides by the count.
#    - Formula: Mean = Sum of all values / Number of values
#    - Example: Mean of [2, 4, 6] = (2+4+6)/3 = 4
#    - numeric_only=True : Only calculates for numerical columns (ignores Species).
#    - Sensitive to outliers (extreme values can pull the mean up or down).

# 2. MEDIAN (Middle Value)
#    df.median(numeric_only=True)
#    - The middle value when data is sorted in order.
#    - If even number of values, median = average of two middle values.
#    - Example: Median of [1, 3, 5, 7, 9] = 5
#    - NOT sensitive to outliers (better than mean for skewed data).

# 3. MODE (Most Frequent Value)
#    df.mode()
#    - The value that appears most often in the data.
#    - Example: Mode of [1, 2, 2, 3, 4] = 2
#    - A dataset can have more than one mode (bimodal, multimodal).
#    - Useful for both numerical and categorical data.
#    - Note: df.mode() works on all columns including Species.

# --------------------------------------------------------------------------
# MEASURES OF VARIABILITY / DISPERSION
# (These tell us how SPREAD OUT the data is)
# --------------------------------------------------------------------------

# 4. STANDARD DEVIATION
#    df.std(numeric_only=True)
#    - Measures how much values deviate (differ) from the mean on average.
#    - Low std  -> values are close to the mean (less spread).
#    - High std -> values are far from the mean (more spread).
#    - Formula: std = sqrt(variance)
#    - Same unit as the original data (e.g., cm).

# 5. VARIANCE
#    df.var(numeric_only=True)
#    - Average of the squared differences from the mean.
#    - Formula: variance = sum((x - mean)^2) / (n - 1)  [sample variance]
#    - Higher variance = more spread out data.
#    - Unit is squared (e.g., cm^2), so std is preferred for interpretation.
#    - Relationship: variance = std^2

# 6. MINIMUM VALUE
#    np.min(df.select_dtypes(include='number'))
#    - Returns the smallest value in each numerical column.
#    - select_dtypes(include='number') selects only numerical columns.
#    - np.min() then finds the minimum of each selected column.

# 7. MAXIMUM VALUE
#    np.max(df.select_dtypes(include='number'))
#    - Returns the largest value in each numerical column.
#    - Same approach as minimum but finds the highest value.

# --------------------------------------------------------------------------
# QUANTILES / PERCENTILES
# (These divide the data into equal parts)
# --------------------------------------------------------------------------

# 8. 1st QUARTILE - Q1 (25th Percentile)
#    df.quantile(0.25, numeric_only=True)
#    - 25% of the data falls BELOW this value.
#    - Also called the lower quartile.

# 9. 2nd QUARTILE - Q2 (50th Percentile)
#    df.quantile(0.50, numeric_only=True)
#    - 50% of the data falls BELOW this value.
#    - This is the same as the MEDIAN.

# 10. 3rd QUARTILE - Q3 (75th Percentile)
#     df.quantile(0.75, numeric_only=True)
#     - 75% of the data falls BELOW this value.
#     - Also called the upper quartile.

# INTERQUARTILE RANGE (IQR) - Important concept:
#     IQR = Q3 - Q1
#     - Measures the spread of the middle 50% of data.
#     - Used to detect outliers:
#       Values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR are outliers.

# --------------------------------------------------------------------------
# SUMMARY TABLE - ALL MEASURES AT A GLANCE
# --------------------------------------------------------------------------
# | Measure            | Function                        | What it tells us         |
# |--------------------|----------------------------------|---------------------------|
# | Mean               | df.mean(numeric_only=True)       | Average value             |
# | Median             | df.median(numeric_only=True)     | Middle value              |
# | Mode               | df.mode()                        | Most frequent value       |
# | Std Deviation      | df.std(numeric_only=True)        | Spread around mean        |
# | Variance           | df.var(numeric_only=True)        | Squared spread            |
# | Minimum            | np.min(df.select_dtypes(...))    | Smallest value            |
# | Maximum            | np.max(df.select_dtypes(...))    | Largest value             |
# | Q1 (25%)           | df.quantile(0.25)                | Lower quartile            |
# | Q2 (50%) = Median  | df.quantile(0.50)                | Middle quartile           |
# | Q3 (75%)           | df.quantile(0.75)                | Upper quartile            |

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# Central Tendency  : A single value that represents the center of the data.
# Variability       : How spread out or scattered the data values are.
# Outlier           : A value that is very different from the rest of the data.
# Percentile        : Value below which a given % of data falls.
# Quartile          : Divides data into 4 equal parts (Q1, Q2, Q3).
# IQR               : Interquartile Range = Q3 - Q1 (middle 50% spread).
# numeric_only=True : Tells pandas to only calculate for number columns.
# select_dtypes()   : Selects columns of a specific data type.

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is Descriptive Statistics?
# A: It is a branch of statistics that summarizes and describes the main
#    features of a dataset using measures like mean, median, mode,
#    standard deviation, etc. It does not make predictions.

# Q: What dataset is used and what does it contain?
# A: The Iris dataset with 150 rows and 5 columns - SepalLengthCm,
#    SepalWidthCm, PetalLengthCm, PetalWidthCm, and Species (3 flower types).

# Q: What is the difference between Mean, Median, and Mode?
# A: Mean = average of all values.
#    Median = middle value when sorted.
#    Mode = most frequently occurring value.

# Q: When should we use Median instead of Mean?
# A: When the data has outliers or is skewed. Median is not affected by
#    extreme values, so it gives a better picture of the center.

# Q: What is the difference between Variance and Standard Deviation?
# A: Variance is the average of squared differences from the mean.
#    Standard Deviation is the square root of variance.
#    Std is preferred because it has the same unit as the original data.

# Q: What are Quartiles?
# A: Quartiles divide data into 4 equal parts.
#    Q1 = 25th percentile (lower quartile)
#    Q2 = 50th percentile (median)
#    Q3 = 75th percentile (upper quartile)

# Q: What is IQR and how is it used?
# A: IQR = Q3 - Q1. It measures the spread of the middle 50% of data.
#    It is used to detect outliers:
#    Values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR are considered outliers.

# Q: What does numeric_only=True do?
# A: It tells pandas to only calculate statistics for numerical columns
#    and ignore text/categorical columns like Species.

# Q: What does select_dtypes(include='number') do?
# A: It selects only the columns that have numerical data types (int or float),
#    ignoring string/object columns. Used before np.min() and np.max().
# =============================================================================
```
