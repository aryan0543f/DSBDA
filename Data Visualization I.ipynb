```python
# Handle missing values
df["Cabin"] = df["Cabin"].replace(to_replace=np.nan, value="unknown")
df["Age"] = df["Age"].interpolate()
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
```

```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM - DATA VISUALIZATION I
# SEABORN PLOTS ON TITANIC DATASET
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS DATA VISUALIZATION?
# --------------------------------------------------------------------------
# Data Visualization is the graphical representation of data.
# It helps us understand patterns, trends, and relationships in data
# that are difficult to see in raw numbers.
# Good visualization makes data easy to interpret and communicate.

# --------------------------------------------------------------------------
# DATASET USED - TITANIC DATASET
# --------------------------------------------------------------------------
# The Titanic dataset contains information about passengers on the Titanic.
# Key columns used for visualization:
#   Pclass   : Passenger class (1 = First, 2 = Second, 3 = Third)
#   Age      : Age of the passenger
#   Fare     : Ticket fare paid by the passenger
#   Sex      : Gender of the passenger (male/female)
#   Survived : Whether the passenger survived (0 = No, 1 = Yes)
#   Cabin    : Cabin number (many missing values -> replaced with 'unknown')
#   Embarked : Port of embarkation (S, C, Q)

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# numpy (np)           : For numerical operations.
# pandas (pd)          : For loading and managing data.
# matplotlib.pyplot    : Base plotting library; used for figure size, layout.
# seaborn (sns)        : High-level visualization library built on matplotlib.
#                        Makes beautiful statistical plots with less code.
# %matplotlib inline   : Jupyter magic command to show plots inside the notebook.

# --------------------------------------------------------------------------
# DATA PREPROCESSING (Before Visualization)
# --------------------------------------------------------------------------
# Before plotting, we clean the data:
#   - Cabin: Replace NaN with 'unknown' using replace()
#   - Age  : Fill missing values using interpolate() (linear interpolation)
#   - Embarked: Fill missing values with mode (most common port)
# After this, df.isnull().sum() should show 0 missing values.

# --------------------------------------------------------------------------
# PLOTS USED IN THIS PRACTICAL
# --------------------------------------------------------------------------

# 1. sns.distplot() - Distribution Plot
#    sns.distplot(df['Pclass'])
#    sns.distplot(df['Age'])
#    sns.distplot(df['Age'], bins=40)
#    sns.distplot(df['Age'], bins=20, kde=False, rug=True)
#    sns.distplot(df['Fare'], hist=True)
#
#    - Shows the distribution (spread) of a single numerical column.
#    - By default shows a histogram + KDE (Kernel Density Estimate) curve.
#    - KDE is a smooth curve that estimates the probability distribution.
#    - bins=40 : Divides data into 40 bars in the histogram.
#    - kde=False : Hides the KDE curve, shows only histogram bars.
#    - rug=True  : Adds small tick marks at the bottom showing each data point.
#    - hist=True : Explicitly shows the histogram (default is True).
#    - Useful to see: Is data normally distributed? Is it skewed?

# 2. sns.jointplot() - Joint Plot
#    sns.jointplot(x=df['Age'], y=df['Fare'], kind='scatter')
#    sns.jointplot(x=df['Age'], y=df['Fare'], kind='hex')
#
#    - Shows the relationship between TWO numerical variables.
#    - Center: scatter plot (or hex plot) of the two variables.
#    - Margins: distribution plots for each variable individually.
#    - kind='scatter' : Shows individual data points as dots.
#    - kind='hex'     : Shows density using hexagonal bins.
#                       Darker hex = more data points in that area.
#                       Better than scatter when there are many overlapping points.
#    - Also shows Pearson correlation coefficient (r) and p-value.

# 3. sns.pairplot() - Pair Plot
#    sns.pairplot(df)
#
#    - Creates a grid of scatter plots for EVERY pair of numerical columns.
#    - Diagonal: shows the distribution of each individual column.
#    - Off-diagonal: scatter plots between each pair of columns.
#    - Very useful for exploring relationships between all features at once.
#    - Can add hue='column' to color by category (like Species or Survived).

# 4. sns.barplot() - Bar Plot
#    sns.barplot(x=df['Sex'], y=df['Fare'])
#
#    - Shows the AVERAGE value of a numerical column (y) for each category (x).
#    - Here: shows the average Fare paid by males vs females.
#    - The error bars (vertical lines) show the confidence interval.
#    - Useful for comparing averages across categories.

# 5. sns.countplot() - Count Plot
#    sns.countplot(df['Pclass'])
#
#    - Shows the COUNT (number of occurrences) of each category.
#    - Here: shows how many passengers were in each class (1st, 2nd, 3rd).
#    - Similar to a bar chart but counts instead of averages.
#    - Useful for understanding the distribution of categorical data.

# --------------------------------------------------------------------------
# SUMMARY TABLE - ALL PLOTS
# --------------------------------------------------------------------------
# | Plot         | Function          | What it shows                          |
# |--------------|-------------------|----------------------------------------|
# | distplot     | sns.distplot()    | Distribution of one numerical column   |
# | jointplot    | sns.jointplot()   | Relationship between two columns       |
# | pairplot     | sns.pairplot()    | All pairwise relationships in dataset  |
# | barplot      | sns.barplot()     | Average of y for each category in x   |
# | countplot    | sns.countplot()   | Count of each category                 |

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# Distribution    : How data values are spread across a range.
# KDE             : Kernel Density Estimate - smooth curve showing distribution.
# Histogram       : Bar chart showing frequency of values in bins/ranges.
# bins            : Number of intervals/bars in a histogram.
# rug             : Small tick marks at the bottom showing each data point.
# Scatter Plot    : Dots showing relationship between two numerical variables.
# Hex Plot        : Like scatter but uses hexagons to show density.
# Correlation     : How strongly two variables are related.
# Pairplot        : Grid of scatter plots for all feature pairs.
# Bar Plot        : Shows average of a numerical variable per category.
# Count Plot      : Shows count/frequency of each category.
# Confidence Interval: Range within which the true mean likely falls.
# seaborn         : Python library for statistical data visualization.
# %matplotlib inline: Shows plots directly in Jupyter notebook.

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is Data Visualization?
# A: Data Visualization is the graphical representation of data to help
#    understand patterns, trends, and relationships that are hard to see
#    in raw numbers.

# Q: What dataset is used and what columns are visualized?
# A: Titanic dataset. We visualize Pclass, Age, Fare, and Sex columns
#    using various seaborn plots.

# Q: What is a distplot and what does it show?
# A: distplot shows the distribution of a single numerical column.
#    It combines a histogram (bars) and a KDE curve (smooth line).
#    It helps see if data is normally distributed or skewed.

# Q: What is the difference between kde=True and kde=False in distplot?
# A: kde=True (default) shows the smooth KDE curve over the histogram.
#    kde=False shows only the histogram bars without the curve.

# Q: What is a jointplot and what are the different kinds?
# A: jointplot shows the relationship between two numerical variables.
#    kind='scatter' shows individual data points.
#    kind='hex' shows density using hexagonal bins (better for large data).
#    The margins show the individual distributions of each variable.

# Q: What is a pairplot?
# A: pairplot creates a grid of scatter plots for every pair of numerical
#    columns. The diagonal shows each column's distribution.
#    It helps explore all relationships in the dataset at once.

# Q: What is the difference between barplot and countplot?
# A: barplot shows the AVERAGE of a numerical variable (y) for each category (x).
#    countplot shows the COUNT (number of rows) for each category.

# Q: What does rug=True do in distplot?
# A: It adds small vertical tick marks at the bottom of the plot,
#    one for each data point. It shows the exact location of each value.

# Q: Why do we use seaborn instead of matplotlib?
# A: Seaborn is built on top of matplotlib and provides higher-level
#    functions that create beautiful statistical plots with less code.
#    It also integrates well with pandas DataFrames.
# =============================================================================
```
