```python
x = df.iloc[:, 0:4].values
x
```

```python
y = df.iloc[:, 4].values
y
```

## # Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

## # Splitting data into 70% training and 30% testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

## # Initializing Naive Bayes Model and training it

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

## # Prediction

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, prediction)
cm
```

```python
accuracy = accuracy_score(y_test, prediction) * 100
print('Accuracy : ', accuracy)
```

```python
print(classification_report(y_test, prediction))
```

```python
# =============================================================================
# IMPORTANT NOTES FOR ORAL EXAM - DATA ANALYTICS III
# NAIVE BAYES CLASSIFIER
# =============================================================================

# --------------------------------------------------------------------------
# WHAT IS THIS PRACTICAL ABOUT?
# --------------------------------------------------------------------------
# This practical is about the Naive Bayes Classifier - a supervised machine
# learning algorithm used for CLASSIFICATION.
# We use the Iris dataset to classify flowers into 3 species based on
# their sepal and petal measurements.
# We also do data visualization using seaborn plots.

# --------------------------------------------------------------------------
# WHAT IS NAIVE BAYES?
# --------------------------------------------------------------------------
# Naive Bayes is a classification algorithm based on Bayes' Theorem.
# It is called 'Naive' because it assumes all features are INDEPENDENT
# of each other (which is rarely true in real life, but works well in practice).
#
# Bayes' Theorem:
#   P(Class | Features) = P(Features | Class) * P(Class) / P(Features)
#
# In simple words:
#   Given the measurements of a flower, what is the probability it belongs
#   to each species? The class with the highest probability is the prediction.
#
# GaussianNB:
#   - The 'Gaussian' version assumes features follow a Normal (bell-curve)
#     distribution.
#   - Used when features are continuous numbers (like sepal/petal measurements).
#
# Advantages of Naive Bayes:
#   - Very fast and simple.
#   - Works well even with small datasets.
#   - Handles multi-class classification easily.

# --------------------------------------------------------------------------
# DATASET USED - IRIS DATASET
# --------------------------------------------------------------------------
# 150 rows x 5 columns.
# 3 classes (species): Iris-setosa, Iris-versicolor, Iris-virginica.
# 50 samples per class.
#
# Columns (after renaming):
#   SepalLengthCm : Length of sepal in cm  -> input feature
#   SepalWidthCm  : Width of sepal in cm   -> input feature
#   PetalLengthCm : Length of petal in cm  -> input feature
#   PetalWidthCm  : Width of petal in cm   -> input feature
#   Species       : Type of flower         -> target (what we predict)

# --------------------------------------------------------------------------
# LIBRARIES USED
# --------------------------------------------------------------------------
# pandas (pd)          : For loading and exploring data.
# numpy  (np)          : For numerical operations.
# matplotlib.pyplot    : For creating plots and figures.
# seaborn (sns)        : For advanced statistical visualizations.
# sklearn.preprocessing: LabelEncoder to convert text labels to numbers.
# sklearn.model_selection: train_test_split to split data.
# sklearn.naive_bayes  : GaussianNB - the Naive Bayes model.
# sklearn.metrics      : confusion_matrix, accuracy_score, classification_report.

# --------------------------------------------------------------------------
# DATA VISUALIZATION (Seaborn Plots)
# --------------------------------------------------------------------------

# 1. sns.relplot() - Scatter Plot
#    sns.relplot(data=df, x='SepalWidthCm', y='SepalLengthCm', hue='Species')
#    - Plots a scatter graph of two features.
#    - hue='Species' colors each point by its species.
#    - Helps see if species are separable based on those two features.
#    - Done twice: once for Sepal, once for Petal measurements.

# 2. sns.pairplot() - Pair Plot
#    sns.pairplot(df, hue='Species')
#    - Plots scatter plots for EVERY pair of features in one grid.
#    - Diagonal shows the distribution (histogram) of each feature.
#    - Very useful to see which feature pairs best separate the 3 species.

# 3. sns.boxplot() - Box Plot
#    sns.boxplot(data=df, x='Species', y='PetalLengthCm')
#    - Shows distribution of a feature for each species.
#    - Box shows Q1, median, Q3. Whiskers show min/max. Dots are outliers.
#    - plt.figure(figsize=(15,10)) sets the figure size.
#    - plt.subplot(2,2,1) divides the figure into a 2x2 grid, uses slot 1.
#    - 4 boxplots are created: one for each measurement column.

# 4. df.corr(numeric_only=True) - Correlation Matrix
#    - Calculates the correlation between all numerical columns.
#    - Values range from -1 to +1:
#        +1 = perfect positive correlation (both increase together)
#        -1 = perfect negative correlation (one increases, other decreases)
#         0 = no correlation

# 5. sns.heatmap() - Heatmap
#    sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='f')
#    - Visualizes the correlation matrix as a colored grid.
#    - annot=True : Shows the correlation values inside each cell.
#    - fmt='f'    : Formats numbers as floating point.
#    - Darker color = stronger correlation.

# --------------------------------------------------------------------------
# STEP 1 - SELECTING FEATURES AND TARGET
# --------------------------------------------------------------------------
# x = df.iloc[:, 0:4].values
#   - Selects columns 0 to 3 (all 4 measurement columns) as input features.
#   - 0:4 means index 0, 1, 2, 3 (4 is excluded).
#
# y = df.iloc[:, 4].values
#   - Selects column 4 (Species) as the target.
#   - y contains text values: 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'.

# --------------------------------------------------------------------------
# STEP 2 - LABEL ENCODING
# --------------------------------------------------------------------------
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# y = le.fit_transform(y)
#
# WHY DO WE NEED LABEL ENCODING?
#   Machine learning models only understand numbers, not text.
#   LabelEncoder converts text categories to integers:
#     'Iris-setosa'     -> 0
#     'Iris-versicolor' -> 1
#     'Iris-virginica'  -> 2
#
# fit_transform() : Learns the mapping AND applies it in one step.

# --------------------------------------------------------------------------
# STEP 3 - SPLITTING DATA
# --------------------------------------------------------------------------
# train_test_split(x, y, test_size=0.3, random_state=0)
#   - 70% training, 30% testing.
#   - Returns: x_train (105 rows), x_test (45 rows), y_train, y_test.

# --------------------------------------------------------------------------
# STEP 4 - TRAINING THE MODEL
# --------------------------------------------------------------------------
# model = GaussianNB()
#   - Creates a Gaussian Naive Bayes model.
#   - Gaussian = assumes features follow a normal distribution.
#
# model.fit(x_train, y_train)
#   - Trains the model. It learns the mean and variance of each feature
#     for each class (species).

# --------------------------------------------------------------------------
# STEP 5 - PREDICTION AND EVALUATION
# --------------------------------------------------------------------------
# prediction = model.predict(x_test)
#   - Predicts the species for each flower in the test set.

# confusion_matrix(y_test, prediction)
#   - 3x3 matrix (since 3 classes).
#   - Diagonal values = correct predictions.
#   - Off-diagonal values = wrong predictions.

# accuracy_score(y_test, prediction) * 100
#   - Percentage of correct predictions.
#   - Here accuracy = 100% because Iris is a clean, well-separated dataset.

# classification_report(y_test, prediction)
#   - Prints a detailed report for each class showing:
#       precision : TP / (TP + FP)
#       recall    : TP / (TP + FN)
#       f1-score  : 2 * (precision * recall) / (precision + recall)
#       support   : Number of actual samples in each class
#
# F1-SCORE:
#   - Harmonic mean of Precision and Recall.
#   - Best when you need a balance between Precision and Recall.
#   - F1 = 1.0 means perfect precision and recall.

# --------------------------------------------------------------------------
# KEY TERMS TO REMEMBER
# --------------------------------------------------------------------------
# Naive Bayes        : Probabilistic classifier based on Bayes' Theorem.
# GaussianNB         : Naive Bayes for continuous features (normal distribution).
# Bayes' Theorem     : P(Class|Features) = P(Features|Class)*P(Class)/P(Features)
# 'Naive' assumption : All features are independent of each other.
# LabelEncoder       : Converts text labels to integers (0, 1, 2...).
# fit_transform()    : Learns AND applies transformation in one step.
# Correlation        : Measures how strongly two features are related (-1 to +1).
# Heatmap            : Color-coded grid to visualize correlation matrix.
# Pairplot           : Grid of scatter plots for all feature pairs.
# Boxplot            : Shows distribution, median, quartiles, and outliers.
# F1-Score           : Harmonic mean of Precision and Recall.
# classification_report: Detailed per-class precision, recall, f1, support.
# plt.subplot(r,c,n) : Divides figure into r rows, c columns, uses slot n.
# plt.figure(figsize): Sets the width and height of the figure in inches.

# --------------------------------------------------------------------------
# POSSIBLE ORAL EXAM QUESTIONS & ANSWERS
# --------------------------------------------------------------------------
# Q: What is Naive Bayes?
# A: Naive Bayes is a supervised classification algorithm based on Bayes'
#    Theorem. It predicts the class with the highest probability given the
#    input features. It is 'naive' because it assumes all features are
#    independent of each other.

# Q: What is GaussianNB and why is it used here?
# A: GaussianNB is the Gaussian (Normal distribution) version of Naive Bayes.
#    It is used when features are continuous numbers (like sepal/petal lengths).
#    It assumes each feature follows a bell-curve (normal) distribution.

# Q: What dataset is used and what are we predicting?
# A: Iris dataset with 150 rows and 5 columns. We predict the species of a
#    flower (Setosa, Versicolor, Virginica) based on 4 measurements.

# Q: Why do we use LabelEncoder?
# A: The Species column has text values ('Iris-setosa', etc.).
#    Machine learning models only work with numbers, so LabelEncoder
#    converts them to integers: Setosa=0, Versicolor=1, Virginica=2.

# Q: What is a Pairplot?
# A: A pairplot shows scatter plots for every combination of feature pairs.
#    The diagonal shows the distribution of each feature.
#    It helps identify which features best separate the classes.

# Q: What is a Heatmap and what does it show?
# A: A heatmap is a color-coded grid that visualizes the correlation matrix.
#    It shows how strongly each pair of features is related.
#    Values close to +1 or -1 mean strong correlation; close to 0 means weak.

# Q: What is the F1-Score?
# A: F1-Score is the harmonic mean of Precision and Recall.
#    Formula: F1 = 2 * (Precision * Recall) / (Precision + Recall)
#    It is useful when you need a balance between precision and recall.
#    F1 = 1.0 means perfect performance.

# Q: What does classification_report show?
# A: It shows per-class precision, recall, f1-score, and support (sample count).
#    Also shows macro avg and weighted avg across all classes.

# Q: Why is accuracy 100% in this practical?
# A: The Iris dataset is very clean and well-separated. The 3 species have
#    distinct enough measurements that Naive Bayes can perfectly classify them.

# Q: What is the difference between Naive Bayes and Logistic Regression?
# A: Both are classification algorithms.
#    Logistic Regression uses the sigmoid function and finds a decision boundary.
#    Naive Bayes uses probability (Bayes' Theorem) and assumes feature independence.
#    Naive Bayes is faster and works well with small data.
#    Logistic Regression is better when features are correlated.
# =============================================================================
```
